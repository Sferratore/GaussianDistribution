```python
from scipy.stats import norm
```

```python
import numpy as npy
```

```python
import matplotlib.pyplot as pyplt
```

```python
x = npy.arange(-5,5,0.1)
```

```python
y = norm.pdf(x) # As default the avg is 0 and variance is 1, so it is treated as a normal distribution
```

```python
pyplt.plot(x, y)
```

```python
y = norm.pdf(x, loc=1, scale=0.5) # We can also choose to change the value of avg (which corresponds to loc) and variance (scale is the standard deviation, square root of variance)
```

```python
pyplt.plot(x,y)
```

```python
eventResults = npy.random.normal(size = 2000, loc=1, scale=0.5) # Generating 2000 values randomically. These are points of a gaussian distribution with avg = 1 and standard deviation = 0.5. If you don't specify loc and scale, default is 0 and 1: normal distr. The points represent results of events which are generated standing at the gaussian probability rate.
```

```python
pyplt.hist(eventResults) # Graphic representing the results. The more the eventResults, the more it will adapt to the shape of a gaussian bell.
```

```python
y = norm.cdf(x, loc=1, scale=0.5) #Cumulative version of the graph
pyplt.plot(x,y)
```
